```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(15, 15))
n_size = 150
e_color = "#a5a5a5"
nx.draw(G, pos, ax=axs[0, 0], node_size=n_size, edge_color=e_color)
nx.draw(G, pos, ax=axs[0, 1], node_size=n_size, edge_color=e_color,
        node_color=list(nx.centrality.degree_centrality(G).values()) )
axs[0, 1].set_title("degree")

nx.draw(G, pos, ax=axs[1, 0], node_size=n_size, edge_color=e_color,
        node_color=list(nx.centrality.closeness_centrality(G).values()) )
axs[1, 0].set_title("closeness")

nx.draw(G, pos, ax=axs[1, 1], node_size=n_size, edge_color=e_color,
        node_color=list(nx.centrality.betweenness_centrality(G).values()) )
axs[1, 1].set_title("betweenness");
```

```python
from networkx.drawing.nx_agraph import graphviz_layout
```

```python
b_centrality = list(nx.centrality.betweenness_centrality(G).values())

fig, axs = subplots(2, 2, figsize=(18, 15))

nx.draw(G, pos=graphviz_layout(G, prog="dot"), 
        node_color=b_centrality, node_size=n_size,
        edge_color=e_color, ax=axs[0, 0])
axs[0, 0].set_title("dot")

nx.draw(G, pos=graphviz_layout(G, prog="circo"), 
        node_color=b_centrality, node_size=n_size,
        edge_color=e_color, ax=axs[0, 1])
axs[0, 1].set_title("circo")

nx.draw(G, pos=graphviz_layout(G, prog="neato"), 
        node_color=b_centrality, node_size=n_size,
        edge_color=e_color, ax=axs[1, 0])
axs[1, 0].set_title("neato")

nx.draw(G, pos=graphviz_layout(G, prog="twopi"), 
        node_color=b_centrality, node_size=n_size,
        edge_color=e_color, ax=axs[1, 1])
axs[1, 1].set_title("twopi");
```
